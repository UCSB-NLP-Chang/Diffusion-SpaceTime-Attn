```python
from bounding_box import bounding_box as bb
import cv2
import numpy as np
import urllib
import cv2
# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
    url_0 = 'https://cs.stanford.edu/people/rak248/VG_100K/'
    url_1 = 'https://cs.stanford.edu/people/rak248/VG_100K_2/'
    try:
      with urllib.request.urlopen(url_0+url) as url:
        resp = url.read()
    except:
      with urllib.request.urlopen(url_1+url) as url:
        resp = url.read()
    image = np.asarray(bytearray(resp))
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    return image
  
gt_idx = 0

data = train_json[gt_idx]

# print(data['objects'])
# data['objects']
pairs = []
objs = []
objs_box = []
for rel in data['relationships']:
  tmp = []
  tmp.append(data['objects'][rel['sub_id']]['class']+str(rel['sub_id']))
  tmp.append(rel['predicate'])
  tmp.append(data['objects'][rel['obj_id']]['class']+str(rel['obj_id']))
  pairs.append(tmp)
for obj in data['objects']:
  objs.append(obj['class'])
  objs_box.append(obj['box'])
print(data['path'])
print(data['id'])
print(pairs)

# draw image
image = url_to_image(data['path'])
for i in range(len(objs_box)):
    bb.add(image, objs_box[i][0], objs_box[i][1], objs_box[i][2], objs_box[i][3], str(objs[i]))
fig=plt.figure(figsize=(15, 15))
plt.imshow(image)
```

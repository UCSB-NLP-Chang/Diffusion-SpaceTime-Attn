```python
import json
import matplotlib.pyplot as plt
rel_dict_path = '../saved/coco_F_seq2seq_v9/test_rel/rel_prior.json'
with open(rel_dict_path, 'r') as fp:
  rel_dict = json.load(fp)
```

```python
rel_dict.keys()
```

```python
fig, ax = plt.subplots()
key = 'surrounding'
fig.set_size_inches(10, 10)
for idx, color in enumerate(['blue', 'orange', 'green', 'black', 'red', 'yellow']):
  key = list(rel_dict.keys())[idx]
  xy_list = rel_dict[key]
  x = []
  y = []
  for i in range(len(xy_list[:100])):
    x.append(xy_list[i][0])
    y.append(-1*xy_list[i][1])
  ax.scatter(x, y, c=color, label=key,
                 alpha=0.5, edgecolors='black')
  plt.xlim(-0.4, 0.4)
  plt.ylim(-0.5, 0.5)
  plt.title("Relation prior distribution")
  plt.xlabel("Δx", fontsize=20)
  plt.ylabel("Δy", fontsize=20)
  plt.legend(fontsize=20)
  plt.savefig('./dawdawd.png')
  plt.grid(True)
```
